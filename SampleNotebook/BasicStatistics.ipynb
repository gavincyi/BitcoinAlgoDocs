```python
import pandas as pd
```

```python
filename = 'exch_quoine_btcusd_snapshot_20170407.csv'
data = pandas.read_csv(filename)
```

```python
data['order_date_time'] = pd.to_datetime(data['order_date_time'], format='%Y-%m-%dD%H:%M:%S.%f000')
data['trades_date_time'] = pd.to_datetime(data['trades_date_time'], format='%Y-%m-%dD%H:%M:%S.%f000')
data
```

```python
trades = data.loc[data['update_type'] == 2, ["trade_px", "trade_volume", "b1", "bq1", "a1", "aq1", "trades_date_time"]]
trades['trades_hour'] = trades['trades_date_time'].dt.hour
trades
```

```python
trades.groupby(['trades_hour'])['trade_volume'].sum()
```
